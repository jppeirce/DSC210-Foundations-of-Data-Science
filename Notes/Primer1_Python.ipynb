## 1.2 Variables: naming a value

**Definition (assignment).** `x = 5` is an *action*: "let the name `x` refer to the value `5`." It is **not** a claim that x equals 5. This is the single most important shift for students coming from math, where `=` states a fact. In Python, `=` *does* something.

````python
# RUN-TOGETHER (the colleague's savings example: a clean first variable)
savings = 100
print(savings)
````

## 1.3 The four core types

Every value has a **type**. Four show up constantly.

**Definition (scalar types).**

| Type | Name | Example | A penguin feature |
| --- | --- | --- | --- |
| `int` | whole number | `100` | number of penguins |
| `float` | decimal | `18.7` | bill depth (mm) |
| `str` | text | `'Adelie'` | species |

The `type()` function tells you which one you have.

````python
# RUN-TOGETHER
print(type(100))          # int
print(type(18.7))         # float
print(type('Adelie'))     # str
print(type(True))         # bool
````

## 2.1 Lists: holding many values

A single variable holds one value. A **list** holds an ordered collection, and a list is our first stand-in for *a column of data*.

**Definition (list).** A list is an ordered, changeable sequence of values, written in square brackets and separated by commas.

````python
# RUN-TOGETHER
# Bill lengths (mm) for the first six penguins in the dataset.
bill_lengths = [39.1, 39.5, 40.3, 36.7, 39.3, 38.9]
print(bill_lengths)
print('how many values:', len(bill_lengths))
````

## 2.2 Indexing and slicing

We reach into a list by **position**. Python counts from **0**, not 1. This trips up every beginner exactly once.

- `x[0]` is the first element, `x[1]` the second.
- `x[-1]` is the last element (negative counts from the end).
- `x[start:stop]` is a **slice**: from `start`, up to but **not including** `stop`.

````python
# RUN-TOGETHER
bill_lengths = [39.1, 39.5, 40.3, 36.7, 39.3, 38.9]
print(bill_lengths[0])    # first  -> 39.1
print(bill_lengths[-1])   # last   -> 38.9
print(bill_lengths[0:3])  # first three -> [39.1, 39.5, 40.3]
````

> **Predict, then run.** The list has 6 elements (positions 0 to 5). Predict each slice before running.

- `bill_lengths[2:5]` -> ____
- `bill_lengths[:2]`  -> ____
- `bill_lengths[3:]`  -> ____

````python
# FILL-IN-LIVE
bill_lengths = [39.1, 39.5, 40.3, 36.7, 39.3, 38.9]
# together: slice out the MIDDLE two values (positions 2 and 3).
middle = bill_lengths[2:4]    # <-- fill in live
print(middle)
````

## 2.3 The copy trap (aliasing)

This is the most common silent bug for new Python users, and it matters in data work because accidentally altering your dataset is hard to notice. Writing `b = a` for a list does **not** make a copy; it makes a second name for the *same* list.

````python
# RUN-TOGETHER
original = [1, 2, 3, 4]
alias = original          # NOT a copy: same list, two names
alias.append(99)

print('alias   :', alias)
print('original:', original)   # surprise: it changed too!
````

> **Discuss.** That is alarming the first time you see it. In your own words, what happened? Why might this be dangerous if `original` were your only copy of a dataset you spent hours collecting?

````python
# RUN-TOGETHER
# The fix: make a real copy with list(...) or a full slice [:]
original = [1, 2, 3, 4]
real_copy = list(original)     # independent copy
real_copy.append(99)

print('real_copy:', real_copy)
print('original :', original)  # untouched, as intended
````

## 2.4 Functions

A **function** is a reusable tool you call by name. You have already used `print()`, `type()`, and `len()`. Functions take **arguments** inside the parentheses and often return a result.

**Definition (function call).** `len(bill_lengths)` *calls* the function `len` with the argument `bill_lengths` and returns its length.

````python
# RUN-TOGETHER
bill_lengths = [39.1, 39.5, 40.3, 36.7, 39.3, 38.9]
print('count  :', len(bill_lengths))
print('largest:', max(bill_lengths))
print('smallest:', min(bill_lengths))
print('sum    :', sum(bill_lengths))
print('rounded avg:', round(sum(bill_lengths) / len(bill_lengths), 2))
````

## 2.5 Methods: functions attached to objects

Almost everything in Python is an **object**, and objects carry their own functions called **methods**. You call a method with a dot: `object.method()`. A list knows how to sort *itself*; a string knows how to uppercase *itself*.

````python
# RUN-TOGETHER
species = 'adelie'
print(species.capitalize())   # str method -> 'Adelie'
print(species.upper())        # -> 'ADELIE'

nums = [3, 1, 4, 1, 5]
print(nums.count(1))          # list method: how many 1s? -> 2
print(nums.index(4))          # position of first 4 -> 2
````

> **Discuss.** `len(nums)` is a **function** but `nums.count(1)` is a **method**. Both do something to `nums`. What is the difference in how you *write* them, and why might `count` "belong to" the list while `len` does not? (There is no single right answer; reason it out.)

## 3.1 Why packages?

Python's core is deliberately small. Specialized tools live in **packages** you load when needed. The data science stack has a few you will use constantly:

| Package | Gives us | Used in |
| --- | --- | --- |
| `numpy` | fast numeric arrays | Primer 2 |
| `pandas` | labeled data tables | the EXPLORE unit |
| `matplotlib` / `seaborn` | plots | Primer 3 |

**Definition (import).** `import` loads a package. `import numpy as np` loads NumPy and gives it the short alias `np`, a near-universal convention.

````python
# RUN-TOGETHER
import math               # a standard-library package (no install)
print(math.pi)            # access pi with dot notation
print(math.sqrt(2))       # square root function
````

### Class Example 3.1: three ways to import (and their trade-offs)

You will see all three of these. They do similar things with different trade-offs.

````python
# RUN-TOGETHER
# 1) import the whole package (clearest, most typing)
import math
print(math.pi)

# 2) import with an alias (the standard for data science)
import numpy as np
print(np.array([1, 2, 3]))

# 3) import one specific name (least typing, hides where it came from)
from math import pi
print(pi)
````

> **Discuss.** Option 3 lets you write just `pi` instead of `math.pi`. Convenient, but what do you *lose*? If you saw `pi` a hundred lines later, would you know which package it came from? When might the explicit `math.pi` be worth the extra typing?

## 3.2 A first taste of the real workflow

Everything so far was hand-typed data. Now the payoff: with two imports we load the actual penguins dataset. This previews the next primers; you are not expected to understand every line, just to see where we are headed.

````python
# RUN-TOGETHER
import seaborn as sns
import numpy as np

penguins = sns.load_dataset('penguins')   # the whole dataset, one line
print('rows, columns:', penguins.shape)
penguins.head()
````

Each row is one penguin (a *case*); each column is a measured *feature*. We pull one feature out as values and summarize it with the tools from Day 2, now applied to real data:

````python
# FILL-IN-LIVE
# together: which column holds flipper length? Fill in the name.
column_name = 'flipper_length_mm'     # <-- fill in live

flippers = penguins[column_name].dropna().tolist()
print('number of measurements:', len(flippers))
print('shortest:', min(flippers), ' longest:', max(flippers))
print('mean:', round(sum(flippers) / len(flippers), 1), 'mm')
````

> **Discuss.** We wrote `.dropna()` to drop missing values before summarizing. Why might a real dataset have missing measurements, and what could go wrong if we forgot to drop them before computing a mean? (We handle missing data carefully in the EXPLORE unit.)
