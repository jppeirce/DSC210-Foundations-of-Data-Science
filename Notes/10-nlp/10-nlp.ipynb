## Regular Expressions


- often shortened to “regex”
- powerful tool for working with text data
- sequence of characters forming a search pattern
- primarily used for pattern matching with strings or string manipulation
- flexible and concise way to identify strings of text such as particular words, patterns of characters, or a combination of these




### Uses of Regular Expressions
Regular expressions are used for several text processing tasks:
- Validation: check if the input data follows a certain format (e.g., email address or a telephone number)
- Search: locate specific strings or substrings within a larger piece of text
- Substitution: replace certain patterns in a string
- Splitting: define the delimiter to split a larger string into a list of smaller substrings
- Data Extraction: for example, scrape web data, where specific patterns need to be extracted from HTML code.



### Pros and Cons of Regular Expressions

Pros:
- Versatile
- Portable

Cons:
- Complexity
- Readability


## Regular Expressions in Python

Built-in `re` module allows us to use regular expressions
- provides several functions, including `match()`, `search()`, `findall()`, `split()`, and `sub()`
- [documentation](https://docs.python.org/3/library/re.html)

### Common Rules

Text, Numbers, Classes
- `|` OR, as "A|B" reads "A OR B"

- `.` any character except “new line” (“\n”)

- `\d` a digit, that is 1, 2, … 9

- `\D` no digit, i.e., any character except 1, 2, … 9

- `\s` a space or tab character

- `\S` no space or tab character, almost all characters

- `[aAbBcC]` any of the contained characters

- `[^aAbBcC]` NONE of the contained characters



### Common Rules

Positions in the String
- `^` beginning of the string

- `$` end of the string

- `\b` word boundary

Repetitions
- `?` zero or once, i.e., the expression is optional

- `+` at least once

- `*` any number of times (including none)

- `{n}` exactly n times

- `{min,max}` at least min times and at most max times

- `{min,}` at least min times

- `{,max}` at most max times

```python
# Example - Regular Expression Building

# Write a Python program that indicates whether a string contains an 'a' followed by a 'b'

def text_match(text):
        patterns = r'a\S*b'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("ac"))
print(text_match("abc"))
print(text_match("abbc"))
print(text_match("adbbc"))
print(text_match("addbbc"))
```

```python
# Example - Regular Expression Building

# Write a Python program that indicates whether a string contains an 'a' directly followed by one or more b's

def text_match(text):
        patterns = 'ab[^b]'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("ac"))
print(text_match("abc"))
print(text_match("abbc"))
```

```python
# Example - Regular Expression Building

# Write a Python program that indicates whether a string contains an 'a' followed by exactly one 'b'

def text_match(text):
        patterns = ' '
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("ac"))
print(text_match("abc"))
print(text_match("abbc"))
```

```python
# Example - Regular Expression Building

# Write a Python program that indicates whether a string contains only sequences of lowercase letters joined with an underscore

def text_match(text):
        patterns = '^[a-z]+_[a-z]+$'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("aab_cbbbc"))
print(text_match("aab_Abbbc"))
print(text_match("Aaab_abbbc"))
print(text_match("_abbbc"))
print(text_match("aaab_abbbC"))
```

```python
# Example - Regular Expression Building

# Write a Python program that matches a string that has an 'a' followed by anything, ending in 'b'.
import re

def text_match(text):
        patterns = r'a.+b$'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("aabbbbd"))
print(text_match("aabAbbbc"))
print(text_match("accddbbjjjb"))
print(text_match("cacddbbjjjb"))
```

```python
# Example - Regular Expression Building

# Write a Python program that matches a word containing 'z'

def text_match(text):
        patterns = '.*z+'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("The quick brown fox jumps over the lazy dog."))
print(text_match("Python Exercises."))
```

```python
# Example - Regular Expression Building

# Write a Python program to extract year, month and date from a url.
# Bonus: extract these as separate elements (capture groups)

def extract_date(url):
        return re.findall(r'(\d{4})/(\d{2})/(\d{2})', url)
url1= "https://www.washingtonpost.com/news/football-insider/wp/2016/09/02/odell-beckhams-fame-rests-on-one-stupid-little-ball-josh-norman-tells-author/"
print(extract_date(url1))
```

```python
# Example - Regular Expression Building

# Power of capture groups!
# Write a Python program to convert a date of yyyy-mm-dd format to dd-mm-yyyy format.

def change_date_format(dt):
        return re.sub(r'(\d{4})-(\d{1,2})-(\d{1,2})', '\\3/\\2/\\1', dt)
dt1 = "2026-01-02"
print("Original date in YYY-MM-DD Format: ",dt1)
print("New date in DD-MM-YYYY Format: ",change_date_format(dt1))
```

## Data Application

How has the popularity of names that contain the letter 'x' changed over time?

Data.gov releases information on baby names, at both national and state-specific levels. `NationalNames.csv` contains data from 1880-2014.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the baby dataset from the specified path
baby = pd.read_csv("/content/drive/MyDrive/Google Drive/Google Drive/Courses/DSC210/Notes/10-nlp/NationalNames.csv")

# Create a new column 'has_x' that is True if the name (in lowercase) contains 'x', otherwise False
baby['has_x'] = baby['Name'].str.lower().str.contains('abcde')

# Calculate the total number of births per year
total_yearly_births = baby.groupby('Year')['Count'].sum()

# Calculate the number of births per year for names containing 'x'
x_names_yearly_births = baby[baby['has_x']].groupby('Year')['Count'].sum()

# Combine the total births and x-name births into a single DataFrame
yearly_counts = pd.DataFrame({
    'total_births': total_yearly_births,
    'x_name_births': x_names_yearly_births
})

# Fill any NaN values in 'x_name_births' with 0, as it means no 'x' names were found for that year
yearly_counts['x_name_births'] = yearly_counts['x_name_births'].fillna(0)

# Calculate the proportion of names containing 'x' for each year
yearly_counts['proportion'] = yearly_counts['x_name_births'] / yearly_counts['total_births']

# Create a line plot to visualize the trend
plt.figure(figsize=(12, 6))
plt.plot(yearly_counts.index, yearly_counts['proportion'], marker='o', linestyle='-')
plt.title('Proportion of Names Containing \'X\' Over Time')
plt.xlabel('Year')
plt.ylabel('Proportion')
plt.grid(True) # Add a grid for better readability
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

### Summary of Observed Trends:

Based on the generated line plot showing the proportion of names containing the letter 'x' over time, the following trends can be observed:

1.  **Initial Stability and Gradual Increase**: From 1880 until around the mid-20th century, the proportion of names containing 'x' remained relatively low and stable, with minor fluctuations.
2.  **Steady Rise**: Starting from the mid-20th century (roughly the 1940s-1960s), there appears to be a gradual, but consistent, increase in the proportion of 'x' names.
3.  **Accelerated Growth in Recent Decades**: The most significant increase in popularity seems to occur from the late 20th century through the early 21st century (e.g., 1980s onwards). The line on the plot shows a steeper upward slope during this period, indicating that names with 'x' have become considerably more popular in recent decades compared to earlier times.
4.  **Overall Upward Trend**: The overall trend is a clear increase in the proportion of names containing the letter 'x' from 1880 to 2014, suggesting a growing preference for names with this letter over the years.

### Tokenization

- **Word-Level Tokenization**: turn a long string into a list of word-like elements you can process one at a time
    - break a cleaned string on spaces to get individual words
    - robust tokenizers handle punctuation attached to words (e.g., "hello!" → ["hello", "!"]) and contractions ("don't" → ["do", "n’t"]).

- **Sentence-Level Tokenization (Optional)**: first identify sentence boundaries (e.g., by looking for periods, question marks, or line breaks) and then tokenize each sentence separately.

- ** Subword or Character-Level Tokenization (Beyond Basics)**: individual words are further split into subword units (e.g., “running” → [“run”, “##ning”]) or even individual characters. This helps handle rare or out-of-vocabulary words.

### Token Cleaning and Normalization

- **Lowercasing** (if not already applied globally)
- **Punctuation Stripping or Filtering**
  - Be careful: sometimes punctuation conveys meaning (e.g., “U.S.A.” or “C++”)
- **Stop-Word Removal (Optional)**: Common words like “the,” “and,” “to,” and “is” often appear so frequently that they contribute little to distinguishing meaning in tasks like topic classification
    - for certain tasks (e.g., sentiment analysis or author attribution), you might retain pronouns and auxiliary verbs because they can carry important signals
- **Digit and Special-Character Handling (Optional)**: Some pipelines remove tokens that contain digits (e.g., “2025” or “A1B2”), replace them with a <NUM> placeholder, or leave them intact if numeric information is relevant (e.g., product codes, dates).
- **Contraction Expansion (If Deferred)**

## One-Hot Encoding

In one-hot encoding, each word in the sentence is represented as a vector in an N-dimensional space
- N is the number of unique words in the sentence
- Each word is represented as a vector of length N
    - all elements being 0, except for one element which is 1, indicating the presence of the word

## Defining Importance

Evaluate how important a word is to a document within a corpus
- importance increases proportionally to the number of times a word appears in the document
- offset by the frequency of the word in the corpus

Here, *document* is some subset of the *corpus* (e.g., sentence vs collection of sentences).




## Term Frequency-Inverse Document Frequency (TF-IDF)

- **Term Frequency (TF)**: $$TF(t)=\frac{\text{Number of times term t appears in a document}}{\text{Total number of terms in the document}}
$$


- **Inverse Document Frequency (IDF)**: $$IDF(t)=\log(\frac{\text{Total number of documents}}{\text{Number of documents with term t in it}})
$$

The **TF-IDF** weight is then the product of these two quantities:

$$\text{TF-IDF}=TF×IDF$$

### The Cost of Context
Using higher n-grams comes at a cost: higher dimensionality

This high dimensionality can lead to two issues:

1. **Sparsity:** Most documents in the corpus will not contain most of the possible bigrams, leading to a feature matrix where most values are zero, i.e., a sparse matrix.
2. **Computational resources:** The computational requirement for storing/processing these feature vectors can become significant, especially for large text collections

Several techniques can help manage this high-dimensionality problem:

- **Feature selection:** We can limit the number of bigrams we include in our feature vector.
- **Dimensionality reduction:** Techniques like Principal Component Analysis (PCA) can be used to reduce the dimensionality of the feature space
- **Using Hashing Vectorizer:** Scikit-learn provides a `HashingVectorizer` that uses a hash function to map the features to indices in the feature vector.
  - constant memory footprint
  - does not require to keep a vocabulary dictionary in memory

```python
def get_nrc_emotion_scores(text):
    if not isinstance(text, str):
        text = str(text) # Ensure input is a string
    emotion_object = NRCLex(text)
    return emotion_object.affect_frequencies

# Apply Emotion Extraction to DataFrame
madrid['nrc_emotion_scores'] = madrid['review_full'].apply(get_nrc_emotion_scores)

# Extract Individual Emotion Scores
# Check an example to get the keys for all possible emotions
# Initialize emotions list with common ones, and add any others found in the first review
example_scores = madrid['nrc_emotion_scores'].iloc[0]
emotions = list(example_scores.keys())

# Create new columns for each emotion
for emotion in emotions:
    madrid[f'nrc_{emotion}'] = madrid['nrc_emotion_scores'].apply(lambda x: x.get(emotion, 0))

# Display the first few rows with the new emotion columns
print("DataFrame with NRC emotion scores:")
print(madrid[['review_full', 'rating_review'] + [f'nrc_{e}' for e in emotions]].head())
```

```python
nrc_emotion_columns = [col for col in madrid.columns if col.startswith('nrc_') and col != 'nrc_emotion_scores']

def get_dominant_nrc_emotion(row):
    emotion_scores = row[nrc_emotion_columns]
    if emotion_scores.empty:
        return 'None'

    # Filter for positive emotion scores
    positive_emotion_scores = emotion_scores[emotion_scores > 0]

    if positive_emotion_scores.empty:
        return 'No_Dominant_Emotion'
    else:
        dominant_emotion_col = positive_emotion_scores.idxmax()
        return dominant_emotion_col.replace('nrc_', '')

madrid['dominant_nrc_emotion'] = madrid.apply(get_dominant_nrc_emotion, axis=1)

print(madrid[['review_full', 'rating_review', 'dominant_nrc_emotion']].head())
```
