## 2. The vocabulary of a dataset

Three terms we will use all semester. Rather than define them in the abstract, we will read them straight off the penguins table.

- **Structured data:** organized in a defined format, typically rows and columns (a spreadsheet, a `.csv`, a database). The penguins table is structured.
- **Unstructured data:** no fixed row/column form: raw text, images, audio, video.

Within a structured (tabular) dataset:

- **Features** are the columns: the properties measured (also called variables or attributes).
- **Cases** are the rows: the units observed (also called records, units, or subjects).
- **Metadata** is data *about* the data: where it came from, when, how it was collected, its units and quality.

## 3. Not all features are the same: measurement scales

A feature's **measurement scale** determines which operations are even meaningful on it. This is a bridge straight from your statistics course.

| Scale | Characteristics | Permissible operations | Example |
| --- | --- | --- | --- |
| **Nominal** | unordered categories | count, mode | penguin species, island |
| **Ordinal** | ordered, unequal gaps | median, percentiles | survey rating (poor/fair/good) |
| **Interval** | equal gaps, no true zero | mean, standard deviation | temperature (C) |
| **Ratio** | equal gaps, true zero | all arithmetic, ratios | flipper length, body mass |

The payoff, stated plainly: **the scale decides which summary is legal.** Averaging a species label is not just bad style; it is a category error.

> **Predict, then run.** Consider two penguin features: `species` (Adelie, Chinstrap, Gentoo) and `body_mass_g`. For which one does asking for a **mean** make sense? Predict, then we test both.

````python
# RUN-TOGETHER
# Ratio feature: a mean is meaningful.
print('mean body mass:', round(penguins['body_mass_g'].mean(), 1), 'g')
````

````python
# RUN-TOGETHER
# Nominal feature: what is the RIGHT summary? Not a mean, but a count.
print(penguins['species'].value_counts())
````

## 4. What makes data "big"?

**Big data:** data that is too large, too complex, or too fast-changing to handle manually or with traditional methods. The penguins table (344 rows) is emphatically **not** big; you could inspect it by hand. Bigness is usually described by a set of **V's**:

- **Volume:** sheer quantity (terabytes, petabytes).
- **Velocity:** how fast it arrives (a sensor stream, a live feed).
- **Variety:** many forms at once (text + images + numbers).
- **Veracity:** how trustworthy or noisy it is.

## 5. Where does the data come from? (GET, up close)

In most classroom examples, including ours today, a dataset appears by magic: `sns.load_dataset('penguins')` and there it is, tidy and documented. Real projects almost never start that way. Huber puts the figure bluntly: a data science project is often **50 to 80% about getting, cleaning, and processing the data**, and it usually begins not with a model, and not even with a table, but with a **question** and then the hunt for data that can actually answer it.

This is the **GET** stage of our loop, and it deserves more respect than beginners give it.

**Definition (data acquisition).** *Data acquisition* is the process of obtaining or creating data suitable for answering a specific question. Note the last three words: the goal is not *some* data, but data that is **relevant, trustworthy, and suitable** for the problem at hand.

Common sources, each with a catch:

| Source | Example | The catch |
| --- | --- | --- |
| Public datasets | Kaggle, UCI, census, weather | "Public" does not mean unbiased, current, or fit for *your* question |
| Web / APIs | a service's official API | APIs are stable; scraping is fragile and sometimes disallowed |
| Internal org data | school records, transaction logs | Privacy, legal limits, access restrictions |
| Scientific data | data published with a paper | May need real domain knowledge to interpret |
| Bought data | a commercial dataset | Paying for it does not make it good |
| Create your own | survey, sensor, experiment | You now own its design, quality, and ethics |

## 6. From data to information to knowledge

Suppose we have acquired our data. Now what? Here is the heart of the matter, and a mild surprise: **data science is not really about data.** Data is the raw material. The actual goal is *understanding*, or more formally, **knowledge**. We dig through data because we hope to *learn* something, not because the data itself is beautiful.

**What does it even mean to "know" something?** A classic answer: you **know** a claim when it is a **justified true belief**, all three at once:

1. you **believe** it,
2. it is **true**, and
3. you have **good reasons** for it.

Why all three? Believing something false is not knowledge. And being *accidentally* right is not either: "I know the next die roll is a 6," which then happens, was a lucky guess, not knowledge, because there was no good reason behind it. **This is the whole job of data science: to supply the justification, the good reasons, that turn a belief into knowledge.** That is what separates an analysis from a hunch.

**Definition (the data - information - knowledge progression).** Building on our two running examples:

| Stage | What it is | Wildlife camera (A) | Customer churn (B) |
| --- | --- | --- | --- |
| **Data** | raw recordings; may be mostly meaningless | a month of footage, mostly empty scenery, wind, shadows | millions of click and purchase logs |
| **Information** | data in context; "data with potential significance for a user" | frames tagged: *animal X, this location, this time* | logs joined to *which* customers cancelled and when |
| **Knowledge** | robust patterns and explanations you can act on | "animal X reliably appears at this clearing at dawn" | "customers churn ~2 months after a price increase" |

Notice the raw data at the top of each column is nearly useless on its own. Value is *added* by context and interpretation as you move down.

**From information to knowledge: reasoning like a detective.** Huber's analogy ties it together. A detective starts with **clues** (a footprint, a broken window), each weak on its own: that is *data*. Connecting and contextualizing the clues yields a coherent picture: that is *information*. Then the detective weighs competing explanations, discards the unlikely ones, and settles on the account that best fits the evidence: that is *knowledge*.

That last move has a name, **inference to the best explanation**:

1. observe the data,
2. propose hypotheses that could explain it,
3. discard hypotheses that do not fit,
4. compare the survivors,
5. select the most plausible, comprehensive explanation.

A related idea from Karl Popper: we rarely *prove* a theory for certain; we make progress by *ruling out* explanations that fail against the data. Your statistics course's logic of rejecting a null hypothesis is exactly this spirit.

**One honest caveat.** The tidy Data -> Information -> Knowledge ladder (often drawn as the **DIKW pyramid**, with Wisdom on top) is a useful teaching picture, but it is also criticized as too simplistic; the real movement from data to knowledge is messy and iterative, not a clean climb. We use it for orientation, not as a law. **Flagged as a judgment call, not settled fact.**

## 7. The workflow, revealed

Step back and look at the road we just traveled this class:

1. We started with a **question** (when is animal X active? why do customers churn?).
2. We **acquired** data suitable for it, and worried about whether it was the *right* data.
3. We turned raw data into **information** by adding context.
4. We reasoned from information toward **knowledge**, an explanation we have good reasons to trust.

That sequence was not improvised. It is *the data science workflow*, and now we can give it its name, the **spine** you have been glimpsing all class.

```
ASK  ->  GET  ->  EXPLORE  ->  MODEL  ->  COMMUNICATE
 ^                                            |
 |________ findings raise new questions _______|
```

- **ASK**: frame the question (Section 5 insisted we start here).
- **GET**: acquire and inspect the data (Section 5).
- **EXPLORE**: clean and contextualize, turning data into information (Section 6, the tagged frames, the joined logs).
- **MODEL**: find and test patterns (Section 6, the dawn-activity pattern, the churn-after-price-increase pattern).
- **COMMUNICATE**: interpret and share the knowledge so someone can act on it.

And the loop: acting on knowledge raises the next question, so we go around again. **Real data science is cyclical, not a straight line.**

**Reconciling with Huber's ASEMIC.** The textbook uses a six-phase version of the same journey, **ASEMIC**. Same road, finer grain. Keep this crosswalk as reference so the two names never confuse you:

| Our loop (the spine) | Huber's ASEMIC | In plain words |
| --- | --- | --- |
| ASK | *(overarching question)* | what are we trying to find out? |
| GET | **A**cquire | obtain and inspect raw data |
| EXPLORE | **S**crub + **E**xplore | clean it; turn data into information |
| MODEL | **M**odel | find/test patterns; build explanations |
| COMMUNICATE | **I**nterpret + **C**ommunicate | decide what it means; share it |

Other standard workflows you will hear about (CRISP-DM, OSEMN) are variations on the same theme. Do not memorize the acronyms; remember the *shape*: question, data, understanding, and a loop back.

- **Data** is recorded observation; it may be **structured** or **unstructured**, with **cases** (rows), **features** (columns), and **metadata** (Sections 1-2).
