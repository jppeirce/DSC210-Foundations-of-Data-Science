## Learning

In machine learning, we use algorithms that **“learn from data”.**
- this means that the rules by which a program decides on its outcome or behavior are no longer hard-coded by a person, but they are automatically searched and optimized.

Examples:
1. Imagine we have the shoe size and the height of many people.
  - We could automatically find a good linear fit function and then use this to make predictions for new data entries.
  - Say, you find footprints of size 12 after a robbery, then your new “machine learning model” can *predict* the height of that person.
2. Think of several chat messages that we clustered into “spam”, “negative”, and “positive”.
  - If we now receive a new message which ends up clearly within the “positive” cluster, then we can obviously risk the *first best guess* of saying that this might be a positive message as well.

## Features and Labels

**Features** are the input variables used to make predictions

**Labels** are the output variable that the model is trying to predict.


## Supervised vs Unsupervised Learning

**Supervised Learning** involves training a model on a labeled dataset
- Each training example is paired with an output label
- The model learns to make predictions based on this input-output mapping.
- Common tasks in supervised learning include *classification* and *regression*.

**Unsupervised Learning** involves training a model on data that does not have labeled responses
- The goal is to infer the natural structure present within a set of data points.
- This is often used for *clustering* and *association* problems.



## Classification vs Regression

**Classification** means that models will predict the class of unknown data points based on what they “learned” from labeled data.
- In a medical diagnosis application, a classification model might predict whether a patient has a particular disease (e.g., positive or negative) based on their medical records.

**Regression** means that models will predict one (or multiple) numerical values for unknown data points.
- We predict house prices based on features like size, location, and the number of bedrooms.


## A Bit More Jargon

- **Model**: a program function
  - It takes a certain, well-defined input (the data, often written as $X$) and generates predictions or labels (written: $y$).
- A model is **trained** on data with known labels.
  - This is what *learning* means in machine learning.
  - Refers to an automated optimization on a given target.
- What is automatically optimized in the model training are the model **parameters**.
  - These are “learned” by the model during training.
  - This is different from general model settings that are called **hyperparameters** and have to be set by us.
- The prediction of labels is called **prediction** or **inference**. This should happen after the model training.



## How Good is Our Model?

Applying supervised machine learning models is tricky!

- We train the model using well-understood data and later want to use its predictions on unknown data.
- We want to use our model for data where we don't know the correct labels ourselves.
- Or, we *could* know the labels but do not want to manually create those for all cases.

Can we really trust the model's predictions?

Everything the model outputs are predictions.
- Every single model output is uncertain and could be wrong.
- Generally, we cannot even estimate the chance of failing.
- We can measure how reliable our model is on average!



## Training vs Testing

For assess model reliability, we need to do (perhaps) the single most important action in machine learning: the **train/test split**.

- For supervised machine learning, we need data ($X$) and corresponding labels ($y$).
- To avoid blindly applying our model to truly unseen and unknown data, we virtually always split our data into training and test data.
  - Training data can be used to train our machine learning model.
  - The test data is kept out of training to assess the quality of predictions of our model.
  - Typically, we want as much data as possible for the training since more data usually correlates with better models. However, we also have to reserve enough data for the test set to later guarantee a meaningful assessment of our model.
- This can easily be done by using the `train_test_split` function from `Scikit-Learn`.

# Common Algorithm: $k$-Nearest Neighbors

$k$-nearest neighbors is, for very good reasons, one of the most commonly known machine learning algorithms.

For any given data point $x$, do the following:
1. Search for the $k$ nearest neighbors within the known data.

2. Predict the label
  - For classification: take the most common label of those data points.
  - For regression: take the mean (or median) of those data points.

That's essentially it!

How is this "learning"?
- The algorithm is clearly a machine learning algorithm because we only define the process. The respective outcomes will be fully dependent on the provided reference data.

## Caveats

- A change in $k$ can lead to entirely different predictions for certain data points.
- Predictions can be highly unstable close to border regions
- Predictions tend to be highly sensitive to the local density of data points
- Predictions can be over-confident: For very unusual inputs or even entirely impossible inputs, the algorithm will still find closest neighbors and make a prediction.

## Distance Metrics

$k$-nearest neighbors is a *distance-based algorithm*
- "closer" means “more similar”
- Must choose a *distance metric*: the measure (or function) by which we calculate the distance between data points

Common Distance Metrics:
- **Euclidean**: straight-line distance between points in feature space, calculated with $$ \sqrt{(a_1-a_2)^2 + (b_1-b_2)^2 + \cdots} = \sqrt{\sum_{i=1}^n (x_{i1} - x_{i2})^2}$$
- **Manhattan**: "taxi-cab" or "city block" metric, calculated with $$ |a_1 - a_2| + |b_1-b_2| + \cdots = \sum_{i=1}^n |x_{i1}-x_{i2}|$$
- **Chebyshev**: "chessboard" distance, calculated with $$ max (|x_{i1}-x_{i2}|) $$
- **Minkowski**: generalizes the three measures above, calculated with $$ \sqrt[p]{\sum_{i=1}^n |x_{i1}-x_{i2}|^p} $$


## Data Scaling

Often, we will simply apply one of the following two scaling methods:

1. **MinMax Scaling** - linearly rescale data such that the lowest occurring value becomes 0 and the highest value becomes 1.
2. **Standard Scaling** - rescale our data such that the mean value will be 0 and the standard deviation will be 1.

*Note:* these methods are used for machine learning algorithms. Usual data inspection (including visualizations) are typically done on the raw data (in original scaling).


## Pros and Cons

Pros
- Can be used for classification and regression
- Very intuitive
- No overfitting
- Does not make impossible predictions

Cons
- Predictions are sensitive to the local density of data points and the choice of $k$
- Can suffer from over-confidence
- Does not scale well for very large datasets

# Common Algorithm - Linear Models

Linear models are a cornerstone of machine learning,

- simple yet powerful approach for both regression and classification tasks
- use a linear combination of input features to predict an output
- favored for their interpretability, computational efficiency, and the foundational role they play in understanding complex machine learning concepts


## Linear Regression

Linear regression is one of the simplest and most fundamental machine learning algorithms. It is used for predictive modeling, specifically for forecasting a quantitative response based on one or more predictor variables (features).

The essence of linear regression lies in fitting a linear equation to the observed data. The equation takes the form: $$ y_i = \beta_0 + \beta_1x_1 + \beta_2x_2 + \cdots + \beta_nx_n + \epsilon $$

- $y_i$ represents the dependent variable we aim to predict
- $x_1, \cdots, x_n$ are the independent variables (features)
- $\beta_0$ is the y-intercept
- $\beta_1, \cdots, \beta_n$ are the coefficients that represent the weight of each feature in predicting the target variable
- $\epsilon$ is the error term, capturing the difference between the observed and predicted values

The goal of linear regression is to find the best-fitting line through the data points that minimizes the *sum of the squared differences between the observed and predicted values*, known as the **least squares criterion**.

## Linear Regression Learning

Linear regression “learns” from the data by adjusting the coefficients ($\beta$ values) to minimize the cost function, typically the mean squared error (MSE) between the predicted and actual values of the dependent variable.
- involves solving for the coefficients that result in the best fit
- can be achieved through various methods, including gradient descent or more direct mathematical approaches like the Ordinary Least Squares (OLS) method.

```python
## Basic Example - Exams

##Generate dummy data

import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

NUM_STUDENTS = 50

# Generate dummy data
np.random.seed(10)
hours = np.random.normal(5, 2, NUM_STUDENTS)  # normally distributed around 5 hours
passed = (hours + 1.5 * np.random.normal(0, 1, NUM_STUDENTS)) > 5  # students passing if hours + noise > 5

hours = np.round(hours, 1)
passed = passed.astype(int)

student_data = pd.DataFrame({
    "exam_preparation_hours": hours,
    "exam_passed": passed
})
student_data.head()

```

## Logistic Regression

- not a traditional regression algorithm -- mostly used for classification
- primarily used when the dependent variable is binary or categorical
- used to model the probability of a certain class or event existing
- models the relationship between a set of independent variables and a dependent variable
- uses a logistic function to output a probability value that is used for classification

## Logistic Regression

The core of logistic regression is the *logistic function*, also known as the *sigmoid function*, defined as $$\sigma(z) = \frac{1}{1+e^{-z}}$$ where $$z = \beta_0 + \beta_1x_1 + \beta_2x_2 + \cdots + \beta_nx_n$$

- $\beta_0$ is the intercept
- $\beta_1, \cdots, \beta_n$ are coefficients
- $x_1, \cdots, x_n$ are the variables
- $e$ is the base of the natural logarithm

The output $\sigma(z)$ ranges between 0 and 1. The probability $P$ of the dependent variable $y$ being 1 is given by $$P(y=1 | x) = \sigma(z) = \frac{1}{1+e^{-(\beta_0 + \beta_1x_1 + \cdots + \beta_nx_n)}} $$

## Logistic Regression - Decision Boundary

The probability output by the logistic function is often compared to a threshold value (commonly 0.5) to make a binary decision.

- If the predicted probability is greater than or equal to 0.5, the output class is 1; otherwise, it is 0
- Can be (and often should be) adjusted based on the specific problem and the cost of false positives and false negatives

## Logistic Regression - Training the Model

Logistic regression is trained using a method called Maximum Likelihood Estimation (MLE). The goal of MLE is to find the values of the coefficients $\beta$ that maximize the likelihood of the observed data. This is done by iteratively adjusting the coefficients to minimize the difference between the predicted probabilities and the actual class labels in the training data.

The **cost function** for logistic regression is the log-loss (or binary cross-entropy), which measures the performance of the classification model:

$$Log-Loss = -\frac{1}{m} \sum_{i=1}^{m} [y_i log(p_i) + (1-y_i)log(1-p_i)]$$

where
- $m$ is the number of observations
- $y_i$ is the actual class label for observation $i$
- $p_i$ is the predicted probability for observation $i$

## Gini Index

- measures the probability of misclassifying a randomly chosen observation if you assign its label by randomly sampling the node's class distribution
- if a node contains K classes and the proportion of class $i$ is $p_i$, the Gini index is $$Gini = 1 - \sum_{i=1}^K p_i^2 $$

## Gini Gain

When evaluating a split, we compute
- the Gini of the parent node
- then the Gini of each child
- the weighted average of those child Gini scores

The Gini Gain is $$\Delta G = Gini_{parent} - (\frac{n_L}{n}Gini_L + \frac{n_R}{n}Gini_R)$$

- $n_L$ and $n_R$ are the sample counts in the left and right child
- $n = n_L + n_R$

The algorithm chooses the split that maximizes this gain

Suppose our root node holds 20 shapes—10 squares and 10 triangles—so each class has proportion 0.5.

## Decision Tree Algorithm

At each step, the tree-growing algorithm:

- Scans each feature's candidate thresholds or category partitions
- Computes parent and child Gini indices, then Gini Gain
- Chooses the split with the largest gain
- Repeats on new nodes until stopping criteria apply.

The most important **stopping criteria** are:
- The reached node only contains datapoints of one sort (so in our example: only squares or triangles)
- The reached node reaches a set minimum size (in Scikit-Learn the `min_samples_leaf` parameter)
- The tree reaches a set maximum depth (in Scikit-Learn the `max_depth` parameter)

## Decision Trees - Pros and Cons

Pros:
- **Interpretability*: Decision trees are easy to understand and visualize, making them useful for communicating results.
- **Versatility**: They can handle both numerical and categorical data without needing feature scaling.
- **Non-linear Relationships**: Capable of capturing complex decision boundaries.

Cons:
- **Overfitting**: Decision trees can easily become too complex and overfit the training data, especially if they are very deep.
- **Instability**: Small changes in the data can result in completely different trees.
- **Bias**: They can be biased towards dominant classes in imbalanced datasets.

### Model Evaluation Metrics - Classification

Confusion Matrices

- Formally, a confusion matrix is an $N x N$ matrix used for evaluating the performance of a classification model, where $N$ is the number of target classes. The matrix compares the actual target values with those predicted by the machine learning model, providing insight into the types of errors made by the classifier.

Elements of a Confusion Matrix:
- True Positives (TP): Correct positive predictions.
- False Positives (FP): Incorrectly predicted positives (Type I error).
- True Negatives (TN): Correct negative predictions.
- False Negatives (FN): Incorrectly predicted negatives (Type II error).

In the fraud detection example, which types of errors are more costly or dangerous?

### Model Evaluation Metrics - Classification

- **Precision (Positive Predictive Value)**: ratio of correctly predicted positive observations to the total predicted positives. $$ \frac{TP}{TP+FP} $$
    - High precision indicates a low false positive rate
    - Good metric when the cost of a false positive is high, such as in email spam classification.

- **Recall (Sensitivity, True Positive Rate)**: ratio of correctly predicted positive observations to all observations in positive class. $$ \frac{TP}{TP+FN} $$
    - Good metric where failing to detect an anomaly can have severe consequences (medical scenarios or fraud detection)

- **F1 Score**: weighted average of Precision and Recall $$ F_1 = 2 * \frac{Precision*Recall}{Precision+Recall}$$
    - takes both false positives and false negatives into account
    - particularly useful when the class distribution is uneven (biased data).

- **Accuracy**: ratio of correctly predicted observations to the total observations $$ Accuracy = \frac{TP+TN}{TP+TN+FP+FN}$$
    - can be misleading in the presence of an imbalanced dataset




### Model Evaluation Metrics - Regression

- **Mean Absolute Error (MAE)**: measures the average magnitude of the errors in a set of predictions, without considering their direction. $$ MAE = \frac{1}{n} \sum_{i=1}^n |y_i-\hat{y_i}|$$ where $y_i$  are the actual values and $\hat{y_i}$ are the predicted values.

- **Mean Squared Error (MSE)** $$ MSE = \frac{1}{n} \sum_{i=1}^n (y_i-\hat{y_i})^2$$
    - MSE is like MAE but squares the difference before summing them all instead of using the absolute value.
    - This has the effect of heavily penalizing larger errors.

## Training and Testing Strategies

1. **Train-Test Split**: divide your data into two sets: the training set and the test set
    - typical ratios are 70% for training and 30% for testing, but these can vary depending on the dataset size.
    - Purpose: Ensures that the model can be evaluated on unseen data.
    - Implementation: `train_test_split`
2. **Train-Validation-Test Split**: divide your data into three sets
    - Purpose: Allows for model tuning (hyperparameters) without touching the test set
    - Implementation: Use `train_test_split` twice!

## Training and Testing Strategies

3. **Cross-Validation**: a robust method for estimating the effectiveness of your model
    - especially useful when dealing with limited data
    - **k-Fold Cross-Validation**:
        - the dataset is divided into `k` smaller sets.
        - the model is trained on `k-1` of these folds, with the remaining part used as the test fold.
        - this process is repeated `k` times, with each of the `k` folds used exactly once as the test set.
    - **Stratified k-Fold**: a variation used for highly imbalanced classes

## Unbalanced Data Techniques

- **Oversampling the Minority Class**: Increasing the number of instances in the minority class by duplicating them to prevent the model from being biased toward the majority class.
- **Undersampling the Majority Class**: Reducing the number of instances in the majority class to balance the dataset.
- **Synthetic Data Generation**: Techniques like SMOTE (Synthetic Minority Over-sampling Technique) generate synthetic samples rather than duplicating existing samples to provide more generalization.

# Scikit-Learn Pipelines

A "Pipeline" is a great tool for encapsulating your entire workflow in machine learning model construction.
  - assemble several steps that can be cross-validated together while setting different parameters
  - helps in maintaining the sequence of transformations and ensuring that all steps in the process are carried out consistently
  - makes the model building process cleaner, easier to manage, and repeatable.

```python
## Pipeline Example Usage

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

# Create a pipeline with a scaler, PCA dimensionality reduction, and logistic regression
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=2)),
    ('logistic_regression', LogisticRegression())
])

# Now the pipeline can be used like a single model
pipeline.fit(X_train, y_train)
predictions = pipeline.predict(X_test)
```
