# Customizing Plots

There are many, many options for plotting data. These include:
- colors
- shapes
- labels
- axes
- many others

Your choices depend on a variety of factors, but boil down to two factors:
- the data
- the story you want to tell with the data

```python
# Example - Basic Plot

plt.plot(year, pop)
plt.show()

# What does this data show?  How do you know that? Would your audience know that?
```

```python
# Example - Add axis labels, title, and some cleanup

plt.plot(year, pop)

# Customize prior to showing the plot

# Add x-axis and y-axis labels, and title the plot
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')

# Change the y-axis tick labels to include 0, and display with 'B' for billions
# First argument is a numeric list, second argument is a str list of same length
plt.yticks([0,2,4,6,8,10,12],
           ['0', '2B', '4B', '6B', '8B', '10B', '12B'])

plt.show()
```

```python
# Example - Gapminder Data

# Basic scatter plot, log scale
plt.scatter(gdp_cap, life_exp)
plt.xscale('log')

# Strings
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)

# Add title
plt.title(title)

# Definition of tick_val and tick_lab
tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']
# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab)


# After customizing, display the plot
plt.show()
```
