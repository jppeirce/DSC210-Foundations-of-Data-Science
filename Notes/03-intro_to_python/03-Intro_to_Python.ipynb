**Definition** The command `x = 5` is an *assignment action*: "let the name `x` refer to the value `5`." It is **not** a claim that x equals 5. This is an important shift from math notation, where `=` states a fact. In Python, `=` *does* something.

## Indexing and slicing

We reach into a list by position or index. Python counts from **0**, not 1.

- `x[0]` is the first element, `x[1]` the second.
- `x[-1]` is the last element (negative counts from the end).
- `x[start:stop]` is a **slice**: from `start`, up to but **not including** `stop`.

````python
# RUN-TOGETHER
bill_lengths = [39.1, 39.5, 40.3, 36.7, 39.3, 38.9]
print(bill_lengths[0])    # first  -> 39.1
print(bill_lengths[-1])   # last   -> 38.9
print(bill_lengths[0:3])  # first three -> [39.1, 39.5, 40.3]
print(bill_lengths[:3]) 

# How do slice out the middle two values:
print(bill_lengths[___:___])  
````

## The copy trap (aliasing)

In data work, accidentally altering your dataset is hard to notice. A common silent bug for new Python users, is creating a alias instead of copy. 

Writing `b = a` for a list does **not** make a copy; it makes a second name for the *same* list.

````python
# RUN-TOGETHER
original = [1, 2, 3, 4]
alias = original          # NOT a copy: same list, two names
alias.append(99)

print('alias   :', alias)
print('original:', original)  

# Keep this in mind especially dealing with raw (original) data
````

````python
# RUN-TOGETHER
# The fix: make a real copy with list(...) or a full slice [:]
original = [1, 2, 3, 4]
real_copy = original.copy()    # independent copy
real_copy.append(99)

print('real_copy:', real_copy)
print('original :', original)  
````

## Importing Packages
Before we can start using modules and methods from packages, we need to import the package (or specific module from the package).

We can do this with an `import` statement, like:

`import numpy`


````python
# Example - Importing Packages
# Use the 'array' function from `numpy`

import numpy
array([1,2,3])

# Note the error behavior here.
# `array()` is a function in the numpy package
# We need to be more specific about referencing that function
````

````python
# Example - Importing Packages
# Use the 'array' function from `numpy`

numpy.array([1,2,3])
# Try print() with this. Are the outputs different?
print(numpy.array([1,2,3]))
````

````python
# Example - Less Typing
import numpy as np
np.array([1,2,3])
````

````python
# Example - Be more selecive

# If we want to use only one function from a package, we can just import that function
# If you do this, you do not need to numpy. (or np.) notation to reference the function
from numpy import array
array([1,2,3])

````

### Which option is the best?

- `import numpy` followed by `numpy.array()` is the clearest version in code, but requires more typing
- `import numpy as np` followed by `np.array()` requires some human memory
- `from numpy import array` followed by `array()` requires the least coding, but hides some detail (requires more human memory)

The structure of `import numpy as np` is used very frequently. There are common shorthand names for many of the packages we'll see.


````python
## Example - Pi
# Let's calculate the circumference and area of a circle with radius r=0.43.
# Circumference = 2*pi*r
# Area = pi*r^2

# The object `pi` doesn't exist in the main distribution of Python, but does exist in the `math` package
# Import the math package
import math

# Calculate C
C = 2 * 0.43 * math.pi

# Calculate A
A = pi * 0.43 ** 2  # `**` is used for exponentiation

print("Circumference: " + str(C))
print("Area: " + str(A))

# Why doesn't this work as it is currently?
````

````python
# Example - Pi
#Common values (like pi) are reasonable objects to import directly

from math import pi
# Calculate C
C = 2 * 0.43 * pi

# Calculate A
A = pi * 0.43 ** 2  # `**` is used for exponentiation

print("Circumference: " + str(C))
print("Area: " + str(A))
````
