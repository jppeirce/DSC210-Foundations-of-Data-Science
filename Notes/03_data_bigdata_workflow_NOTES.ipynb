## 2. The vocabulary of a dataset

Three terms we will use all semester. Rather than define them in the abstract, we will read them straight off the penguins table.

- **Structured data:** organized in a defined format, typically rows and columns (a spreadsheet, a `.csv`, a database). The penguins table is structured.
- **Unstructured data:** no fixed row/column form: raw text, images, audio, video.

Within a structured (tabular) dataset:

- **Features** are the columns: the properties measured (also called variables or attributes).
- **Cases** are the rows: the units observed (also called records, units, or subjects).
- **Metadata** is data *about* the data: where it came from, when, how it was collected, its units and quality.

## 3. Not all features are the same: measurement scales

A feature's **measurement scale** determines which operations are even meaningful on it. This is a bridge straight from your statistics course.

| Scale | Characteristics | Permissible operations | Example |
| --- | --- | --- | --- |
| **Nominal** | unordered categories | count, mode | penguin species, island |
| **Ordinal** | ordered, unequal gaps | median, percentiles | survey rating (poor/fair/good) |
| **Interval** | equal gaps, no true zero | mean, standard deviation | temperature (C) |
| **Ratio** | equal gaps, true zero | all arithmetic, ratios | flipper length, body mass |

The payoff, stated plainly: **the scale decides which summary is legal.** Averaging a species label is not just bad style; it is a category error.

> **Predict, then run.** Consider two penguin features: `species` (Adelie, Chinstrap, Gentoo) and `body_mass_g`. For which one does asking for a **mean** make sense? Predict, then we test both.

````python
# RUN-TOGETHER
# Ratio feature: a mean is meaningful.
print('mean body mass:', round(penguins['body_mass_g'].mean(), 1), 'g')
````

````python
# RUN-TOGETHER
# Nominal feature: what is the RIGHT summary? Not a mean, but a count.
print(penguins['species'].value_counts())
````

## 4. What makes data "big"?

**Big data:** data that is too large, too complex, or too fast-changing to handle manually or with traditional methods. The penguins table (344 rows) is emphatically **not** big; you could inspect it by hand. Bigness is usually described by a set of **V's**:

- **Volume:** sheer quantity (terabytes, petabytes).
- **Velocity:** how fast it arrives (a sensor stream, a live feed).
- **Variety:** many forms at once (text + images + numbers).
- **Veracity:** how trustworthy or noisy it is.

## 5. From data to knowledge

Raw data by itself is often meaningless. A widely used ladder describes how meaning gets added:

- **Data:** raw values or findings. Unprocessed reality; possibly incomprehensible on its own. *(The number `181`.)*
- **Information:** data placed in context, structured and categorized. *(`181` is a flipper length in millimeters.)* Otte et al. (2020): "information is data with potential significance for a user."
- **Knowledge:** understanding built from information, combining facts, theories, and experience. *(Adelie penguins tend to have shorter flippers than Gentoos, and here is why that might be.)*

This is sometimes drawn as a pyramid with **Wisdom** at the top (knowing which questions are worth asking at all).

Huber frames the same progression with a memorable analogy: **the data scientist as detective.**

- Data are **clues**.
- Clues pieced together become a **story** (information).
- Story plus context becomes an **informed conclusion** about the truth (knowledge).

The detective does not passively receive the answer; they *reason* toward the best explanation of the clues.

**Inference to the best explanation** (Bartelborth, 2017) makes that reasoning explicit, and, not coincidentally, it previews our workflow:

1. **Observe** the phenomena (gather the raw data / clues).
2. **Formulate hypotheses** that might explain them.
3. **Refine:** discard hypotheses that do not fit the facts.
4. **Compare:** which surviving hypothesis explains the data best?
5. **Select** the most plausible, comprehensive explanation.

## 6. The data science workflow (the spine, formally)

A **workflow** is a blueprint that guides a project from question to communicated result. Several standard ones exist:

- **CRISP-DM** (circular): Business Understanding, Data Understanding, Data Preparation, Modeling, Evaluation, Deployment.
- **OSEMN** (linear): Obtain, Scrub, Explore, Model, iNterpret.
- **Tree-like** workflows branch after a step into multiple paths.

Huber's text uses a tailored six-phase version, **ASEMIC**:

**A**cquire, **S**crub, **E**xplore, **M**odel, **I**nterpret, **C**ommunicate.

Huber stresses, and we agree, that this is **iterative, not linear**: after modeling you often return to scrubbing; after interpreting you may revisit exploration.

### Reconciling ASEMIC with our course loop

We will use a five-stage loop as the course **spine** because it is easy to hold in your head and it makes the iteration visible. Huber's ASEMIC is the same process at finer grain. Here is the crosswalk, so the two never confuse you:

| Course loop (our spine) | Huber's ASEMIC | In plain words |
| --- | --- | --- |
| **ASK** | *(the overarching question)* | What are we trying to find out? |
| **GET** | **A**cquire | Obtain and inspect the raw data |
| **EXPLORE** | **S**crub + **E**xplore | Clean it; visualize; find patterns |
| **MODEL** | **M**odel | Describe or predict |
| **COMMUNICATE** | **I**nterpret + **C**ommunicate | Decide what it means; share it |

Both are loops. Findings at the end feed new questions at the start. The one idea to carry away: **real data science is cyclical.** You will walk this loop many times before the semester ends.

## 7. First stop on the loop: acquiring data (GET)

Since GET is where every project begins, four questions to ask of any dataset before you trust it:

1. **Type of data needed:** quantitative or qualitative? temporal? spatial? What would actually answer the question?
2. **Quality and quantity:** enough breadth and depth? gaps? inconsistencies?
3. **Sources:** available, accessible, and *trustworthy*?
4. **Usage restrictions:** any legal, commercial, or privacy limits? **Having data is not the same as being allowed to use it.**

Common sources include public data portals (Kaggle, Zenodo, the UCI repository), web scraping (within legal and ethical limits), internal corporate data, academic datasets attached to papers, data on request, and commercial datasets.

## Summary

- **Data** is recorded observation or measurement; it may be **structured** (rows and columns) or **unstructured** (text, images, audio).
- In a table: **cases** are rows, **features** are columns, **metadata** is data about the data.
- A feature's **measurement scale** (nominal / ordinal / interval / ratio) decides which summaries are meaningful.
- **Big data** (Volume, Velocity, Variety, Veracity) is what broke the old tools and motivated data science as a field.
- **Data -> Information -> Knowledge**: meaning is added by context and reasoning; the data scientist reasons like a detective toward the best explanation.
