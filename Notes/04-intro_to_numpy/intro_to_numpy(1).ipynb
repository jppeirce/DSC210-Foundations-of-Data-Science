# Introduction to NumPy (Numeric Python)

Typically in Data Science, you'll often want to carry out operations over entire collections of values, and you want to do this fast. 

With lists, this is a problem!

```python
# Example - BMI

# Suppose we have height and weight data for a collection of people
# Our goal is to calculate BMI for each person
# We can calculate BMI as weight/height^2

height = [1.73, 1.68, 1.71, 1.89, 1.79]
weight = [65.4, 59.2, 63.6, 88.4, 68.7]

bmi = weight/height**2  

# This doesn't work. What's the issue?
```

```python
# NumPy to the Rescue!

# Note, you may need to install numpy on your own system prior to using the package

NumPy provides an alternative to lists: arrays
- Similar to lists, but allow for calculations across the array
- Fast!

NumPy arrays can contain only one type of data
- If you create a "mixed" array, Python will convert the data to a more general type (like `str`)
- NumPy arrays are another type of object (like `str`, `float`, and `list`)
- Methods are associated with arrays, and may behave differently than other methods with the same name
```

```python
# Example - BMI
import numpy as np
np_height = np.array(height) #Note the use of `_` in the name
np_weight = np.array(weight)

bmi = np_weight/np_height**2  
print(bmi)

# Calculations were performed element-wise
print(65.4/1.73**2)
```

```python
# Example - Method Acting
python_list = [1,2,3]   #Standard list
numpy_array = np.array([1,2,3])   #Standard array

print(python_list + python_list)
print(numpy_array + numpy_array)

print(python_list[1])
print(numpy_array[1])
```

```python
# Example - More subsetting approaches

# Suppose we wanted to find all of the BMI values over 24
bmi > 24  #creates a NumPy array of booleans

# Return only the values above 24, using the array of booleans
bmi[bmi > 24]
```

# NumPy Arrays

Let's look at the output `<class 'numpy.ndarry'>`
- The `numpy.` part indicates that the object was defined using the numpy package
- The `ndarry` part stands for 'n dimensional array'

`np_height_in` and `np_weight_lb` are both 1-dimensional arrays



```python
## 2D NumPy Arrays

# Using height and weight data from the bmi example above
np_2d = np.array([[1.73, 1.68, 1.71, 1.89, 1.79],
                  [65.4, 59.2, 63.6, 88.4, 68.7]])

print(np_2d)

# Shape attribute for arrays (provides #rows, #cols)
# Attributes do not get (), but methods do!
print(np_2d.shape)

# Access the element in the first row and third column
print(np_2d[0,2]) # Remember 0 indexing!
print(np_2d[0][2]) # also works!

# Access the elements for the second and third person
np_2d[:,1:3] # : indicates range, with blanks yielding (effectively) "start" and "end+1"
# Remember what 1:3 means: start with the index 1 (second position) and stop before index 3 (4th position)
```
