# 1. Why Arrays?

## 1.1 The issue with lists

Typically in Data Science, you'll often want to carry out operations over entire collections of values, and you want to do this fast.

With lists, this is a problem!

```python
## Example - BMI
# RUN-TOGETHER

# Suppose we have height and weight data for a collection of people
# Our goal is to calculate BMI for each person
# We can calculate BMI as weight/height^2

height = [1.73, 1.68, 1.71, 1.89, 1.79]
weight = [65.4, 59.2, 63.6, 88.4, 68.7]

bmi = weight/height**2

# This doesn't work. What's the issue?
```

 For a **list**, `**` and `/` are not defined element-by-element; there is no way to say "divide each weight by each height squared" without a loop. We want something that just does the math. That is NumPy.

## 1.2 NumPy to the Rescue!

 ** Note, you may need to install numpy on your own system prior to using the package **

NumPy provides an alternative to lists: arrays
- Similar to lists, but allow for calculations across the array
- Fast!

**Definition.** A NumPy *array* (`ndarray`) is an ordered collection of values, like a list, but with two superpowers: (1) math happens **element-wise** across the whole array, and (2) it is fast (the looping runs in compiled C, not Python).

One rule: NumPy arrays can contain only **one** type of data.

- If you create a "mixed" array, Python will convert the data to a more general type (like `str`)
- NumPy arrays are another type of object (like `str`, `float`, and `list`)

```python
## Example - BMI
# RUN-TOGETHER
import numpy as np

height_m = np.array([1.73, 1.72, 1.61, 1.89, 1.79])
weight_kg = np.array([65.4, 59.2, 63.6, 88.4, 68.7])

bmi = weight_kg/height_m**2
print(bmi)

# Calculations were performed element-wise
print(65.4/1.73**2)
```

### 1.4 Broadcasting: array meets scalar

*Broadcasting* is how NumPy applies an operation between an array and a single value: the single value (called a scalar) is applied to every element.

```python
# RUN-TOGETHER
import seaborn as sns
penguins = sns.load_dataset('penguins')
flippers_mm = penguins['flipper_length_mm'].dropna().to_numpy()

flippers_cm = flippers_mm / 10     # every value divided by 10
print('first five (mm):', flippers_mm[:5])
print('first five (cm):', flippers_cm[:5])
```

Comparing an array to a value gives an array of `True`/`False`. Indexing an array *with* that boolean array keeps only the `True` positions. This replaces filtering loops and reads almost like English.

```python
## Example - More subsetting approaches
# RUN-TOGETHER
#
# Recall:
print(bmi)

# Suppose we wanted to find all of the BMI values over 24
print(bmi > 24)  #creates a NumPy array of booleans

print('how many over 24:', sum(bmi>24)) # False = 0, True = 1

# Return only the values above 24, using the array of booleans
#print(bmi[bmi > 24])
```

## 2.1 Two-dimensional arrays

Real data are often organized in a table: many rows, several columns. NumPy represents that as a **2D array**. Now `shape` matters and indexing takes a `[row, column]` pair.

```python
# RUN-TOGETHER
import numpy as np

# rows = [heights (in)], [weights (lb)]
players = np.array([[74, 74, 72, 72, 73, 69, 69, 71],       # heights
                    [180, 215, 210, 210, 188, 176, 209, 200]])  # weights

# size does not work to find the dimensions of n-dimentional arrays
print('size of an array: ', players.size)

# instead we use the shape attribute
#print('shape (rows, cols):', players.shape)
```

In data science, we commonly want our data arranged in a different way than what we see in `players`. We prefer that each row is an observational unit (i.e. player) and each column is a feature (i.e height, weight).

We will need to swap rows and the columns so that the height, weight rows become height, weight columns. This can easily be done by using the transpose attibute.

```python
print('original:\n', players) # \n forces a line break
players_dat = players.T
print('transposed:\n', players_dat)
```

### 2.1 Indexing and slicing of 2D arrays

> array_name[row, column] produces the element in location (row, column)

Note, in Python:
- indexing starts a 0
    - `array_name[0,0]` is the element in first row and first column.
    - `array_name[-1,-1]` is the last element (negative counts from the end).
- `:` produces all values
- `array_name[row]` or `array_name[row, :]`produces the single row indexed by `row`
- `array_name[:, column]` produces the single column indexed by `column`

We can slice vales from rows, from columns, or both using the notation `start:stop`
- Remember: the slice is from `start`, up to but **not including** `stop`.


```python
print(players_dat)

# data for first player
print('player 0:', players_dat[0])

# Access the element in the third row and first column
print(players_dat[2,0]) # Remember 0 indexing!
print(players_dat[2][0]) # also works!

# Access the elements for the forth and fifth person
# Think first:
# start = 3: fouth person is in row 3 because index starts at 0
# end = 5: need one more than where we want to stop
print('fourth and fifth:\n', players_dat[3:5,:])

# boolean mask heights < 70
# Think first:
# first column of heights: players_dat[:,0]
# boolean mask: players_dat[:,0] < 70  -> False False ....
print('height < 70:\n', players_dat[players_dat[:,0] < 70, :] )
```
