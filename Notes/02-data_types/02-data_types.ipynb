## 2. The vocabulary of a dataset

There are two general classifications of data:

- Structured data: organized in a defined format
    - Typically arranged in rows and columns (sometimes called rectangular data)
    - Examples: a spreadsheet, `.csv` files, databases, the penguin dataset
- Unstructured data: no specific form or model
    - Examples: Text documents, images, video, audio, etc.

Within a structured (tabular) dataset:

- Features: distinct properties measured
    - Also called variables or attributes
    - Generally the 'columns' of a tabular dataset
    - In UWL student records, these could be 'Academic Year', 'Major', or 'GPA'
- Cases: unit of observation of a dataset
    - Sometimes called 'units' or 'subjects'
    - Generally the 'rows' of a tabular dataset
    - In UWL student records, these would be the individual students

**Note:** For an image, the cases could be the pixel $(x,y)$ location or a pixel numbering and the features could be the RBG color code, or object/no object.

- Metadata: information about the data
    - context, quality, condition, origin, and other characteristics.
    - If data is a book, metadata would contain information about content, author, and publication details.

## 3. Feature Types 

Determining the type of features we record is the first step in deciding how we visualize the features and directs which anaylsis tool or model that we use.

| Scale Type | Characteristics | Data Type | Operations | Examples |
| ------------ | ------------------------------------------- | ----------- | --------------------------------------------- | ------------------------------- |
| Nominal    | Categories without order                      | Categorical | Classification, Mode                            | penguin species, island, major                    |
| Ordinal    | Ordered categories, unequal intervals         | Categorical | Sorting, Median, Percentiles                    | Ratings (poor, fair, good)        |
| Interval   | Equal intervals, no true zero                 | Numerical   | Addition, Subtraction, Mean, Standard Deviation | Temperature (Celsius, Fahrenheit) |
| Ratio      | Equal intervals, true zero, meaningful ratios | Numerical   | All arithmetic operations                       | Height, Weight, Age, Income       |

## 4. What makes data "big"?

** Definition: ** *Big data* are data that are too large, too complex, or too fast-changing to handle manually or with traditional methods. The penguins table (344 rows) is **not** big; we could inspect it by hand. Bigness is usually described by a set of V's:

- **Volume:** sheer quantity (terabytes, petabytes).
- **Velocity:** how fast it arrives (a sensor stream, a live feed).
- **Variety:** many forms at once (text + images + numbers).

## 5. Where does data come from?

In most classroom examples, including ours today, a dataset appears by magic: `sns.load_dataset('penguins')` and there it is, tidy and documented. Real projects almost never start that way. 

Huber puts the figure bluntly: a data science project is often **50 to 80% about getting, cleaning, and processing the data**, and it usually begins not with a model, and not even with a table, but with a **question** and then the hunt for data that can actually answer it.

**Definition:** *Data acquisition* is the process of obtaining or creating data suitable for answering a specific question. Note the last three words: the goal is not *some* data, but data that is **relevant, trustworthy, and suitable** for the problem at hand.

Common sources, each with a catch:

| Source | Example | The catch |
| --- | --- | --- |
| Public datasets | Kaggle, UCI, census, weather | "Public" does not mean unbiased, current, or fit for *your* question |
| Web / APIs | a service's official API | APIs are stable; scraping is fragile and sometimes disallowed |
| Internal org data | school records, transaction logs | Privacy, legal limits, access restrictions |
| Scientific data | data published with a paper | May need real domain knowledge to interpret |
| Bought data | a commercial dataset | Paying for it does not make it good |
| Create your own | survey, sensor, experiment | You now own its design, quality, and ethics |

## 7. From data to information to knowledge

Suppose we have acquired our data. Now what? Here is the heart of the matter, and a mild surprise: 

> **Data science is not really about data.** 

Data is the raw material. The actual goal is *understanding*, or more formally, **knowledge**. We dig through data because we hope to *learn* something, not because the data itself is beautiful.

**What does it even mean to "know" something?** A classic answer: you **know** a claim when it is a **justified true belief**, all three at once:

1. you **believe** it,
2. it is **true**, and
3. you have **good justification** for it.

Why all three? 

Believing something false is not knowledge. And being *accidentally* right is not either: "I know the next coin flip is heads," which then happens, was a lucky guess, not knowledge, because there was no good reason behind it. 

**This is the whole job of data science: to supply the justification, the good reasons, that turn a belief into knowledge.** 

That is what separates an analysis from a hunch.

**Definition:** The *data - information - knowledge progression:*

| Stage | What it is | Customer churn (A) | Wildlife camera (B) |
| --- | --- | --- | --- |
| **Data** | raw recordings; may be mostly meaningless | millions of click and purchase logs |  a month of footage, mostly empty scenery, wind, shadows |
| **Information** | data in context; "data with potential significance for a user" | logs joined to *which* customers cancelled and when | frames tagged: *animal X, this location, this time* |
| **Knowledge** | robust patterns and explanations you can act on |  "customers churn ~2 months after a price increase" | "animal X reliably appears at this clearing at dawn" |

Notice the raw data at the top of each column is nearly useless on its own. Value is *added* by context and interpretation as you move down.

## 8. From information to knowledge: reasoning like a detective.

A detective starts with **clues** (a footprint, a broken window), each weak on its own: that is *data*. 

Connecting and contextualizing the clues yields a coherent picture: that is *information*. 

Then the detective weighs competing explanations, discards the unlikely ones, and settles on the account that best fits the evidence: that is *knowledge*.

That last move has a name, **inference to the best explanation**:

1. observe the data,
2. propose hypotheses that could explain it,
3. discard hypotheses that do not fit,
4. compare the survivors,
5. select the most plausible, comprehensive explanation.

We rarely *prove* a theory for certain; we make progress by *ruling out* explanations that fail against the data. Does this sound familar?

## 8. Data Science Workflow

A blueprint that can guide a project from inception to completion

Examples include:
- Circular: CRISP-DM: Cross-Industry Standard Process for Data Mining
    - Business Understanding, Data Understanding, Data Preparation, Modeling, Evaluation, and Deployment
- Linear: OSEMN
    - Obtain, Scrub, Explore, Model, and iNterpret
- Tree:
    - after a certain step, the process might branch out into multiple paths, reflecting the multifaceted nature of some data science projects

The Huber text follows a slightly more tailored version: **ASEMIC**
- **A**cquire data
- **S**crub data
- **E**xplore data
- **M**odel data
- **I**nterpret results
- **C**ommunicate results

## The workflow model for this class

```
ASK  ->  GET  ->  EXPLORE  ->  MODEL  ->  COMMUNICATE
 ^                                            |
 |________ findings raise new questions ______|
```

- **ASK**: frame the question.
- **GET**: acquire and inspect the data.
- **EXPLORE**: clean and contextualize, turning data into information.
- **MODEL**: find and test patterns (for example, the churn-after-price-increase pattern or the dawn-activity pattern).
- **COMMUNICATE**: interpret and share the knowledge so someone can act on it.

And the loop: acting on knowledge raises the next question, so we go around again. 

**Real data science is cyclical, not a straight line.**

**Reconciling with Huber's ASEMIC.** The textbook uses a six-phase version of the same journey, **ASEMIC**. Same road, finer grain. 

| Our loop  | Huber's ASEMIC | In plain words |
| --- | --- | --- |
| ASK | *(overarching question)* | what are we trying to find out? |
| GET | **A**cquire | obtain and inspect raw data |
| EXPLORE | **S**crub + **E**xplore | clean it; turn data into information |
| MODEL | **M**odel | find/test patterns; build explanations |
| COMMUNICATE | **I**nterpret + **C**ommunicate | decide what it means; share it |

Other standard workflows you will hear about (CRISP-DM, OSEMN) are variations on the same theme. Do not memorize the acronyms; remember the *shape*: question, data, understanding, and a loop back.
