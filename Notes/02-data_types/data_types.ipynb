# What is/are Data?

Broadly, we can define 'data' as information recorded through observations or measurements. These can come from many sources:

- tally sticks (or bones)
- primary sources in historical text
- London parish death records
- comic book mural locations in Brussels
- wearable device (fitness) data
- social media

*Side note:* there is disagreement about whether 'data' is a singular or a plural noun.


# Data Jargon

- Structured data: organized in a defined format
    - Typically arranged in rows and columns (sometimes called rectangular data)
    - Examples: databases, .csv files, most anything that can be arranged in Excel
- Unstructured data: no specific form or model
    - Text documents, images, video, etc.

- Features: distinct properties of a data set
    - Sometimes called attributes or variables
    - Generally the 'columns' of a tabular dataset
    - In UWL student records, these could be 'Academic Year', 'Major', or 'GPA'
- Cases: unit of observation of a dataset
    - Sometimes called 'units' or 'subjects'
    - Generally the 'rows' of a tabular dataset
    - In UWL student records, these would be the individual students

Note: For an image, the cases could be the pixel $(x,y)$ location or a pixel numbering and the features could be the RBG color code, or object/no object.

- Metadata: information about the data
    - context, quality, condition, origin, and other characteristics.
    - If data is a book, metadata would contain information about content, author, and publication details.

# Data - Information - Knowledge - Wisdom

- Data: The raw values or findings obtained through various means such as observations, measurements, surveys, etc.
    - Unprocessed reality
    - Can be entirely incomprehensible or meaningless
- Information: Data in context
    - Has been processed, categorized, and given structure
    - “Information is data with potential significance for a user” [Otte et al., 2020].
- Knowledge: Understanding the information.
    - encompasses facts, theories, applicability, and rules

# Data

Conceived as being composed of symbols or signs, representing stimuli or signals, that are "of no use until ... in a usable (that is, relevant) form" (Rowley 2007)
- (Fact) Raw facts or observations
- (Stimuli) One or more kinds of energy waves or particles (light, heat, sound, force, electromagnetic) selected by a conscious organism or intelligent agent on the basis of a preexisting frame or inferential mechanism in the organism or agent"
- (Symbol) A property of an object, an event or of their environment



# Information

Data that are endowed with meaning and purpose (Rowley 2007)

- Structural vs Functional Information
- Symbolic vs Subjective Information
    - "the meaning of statements as they are intended by the speaker/writer and understood/misunderstood by the listener/reader"


# Knowledge

"Knowledge is a fluid mix of framed experience, values, contextual information, expert insight and grounded intuition that provides an environment and framework for evaluating and incorporating new experiences and information. It originates and is applied in the minds of knowers. In organizations it often becomes embedded not only in documents and repositories but also in organizational routines, processes, practices and norms." (Wallace 2007)

- (Process) synthesis of multiple sources of information over time
- (Procedure) perception, skills, training, common sense and experience
- (Proposition) "belief structuring" and "internalization with reference to cognitive frameworks" (Rowley 2007)


# Bigger Picture - from Data to Knowledge
It is more useful to relate to the data, information, and knowledge as sets of signs rather than as meaning and its building blocks (Zins, 2009)

Data Scientists are Data Detectives!
- Data are clues (Data)
- Data pieced together are stories (Information)
- Combined data and context provide informed guesses about the truth (Knowledge)

How do we move from data to knowledge (or even wisdom)?

# "Inference to the Best Explanation” method [Bartelborth, 2017]

- Observation: Start by observing phenomena. This is your raw data. (in the detectives picture: gather clues)
- Hypothesis Formulation: Develop potential hypotheses that might explain the observed data.
- Refinement: Some theories might not fit the facts. So, they’re discarded.
- Comparison: Which of our remaining theories best explains the data?
- Selection: Choose the hypothesis (or hypotheses) that offers the most plausible and comprehensive explanation in line with other established knowledge.

# Data Acquisition

Key Considerations:
1. **Type of Data Needed**: What kind of information will help address the problem? Is it quantitative, qualitative, temporal, or spatial?
2. **Data Quality and Quantity**: Do the data offer enough breadth and depth to draw meaningful conclusions? Are there gaps or inconsistencies that might pose challenges?
3. **Data Sources**: Is the data readily available? Is it accessible? And can we trust the source?
4. **Usage Restrictions**: Are there any commercial, legal, or privacy restrictions tied to the data? This determines if we can work with the data at all and if so, under what conditions.


# Data Sources

Common Data Sources:
- **The Internet I**: Public datasets are easily accessible and can be sourced from governmental agencies, NGOs, or platforms like [Kaggle](https://www.kaggle.com/), [Zenodo](https://zenodo.org/), and [UCI](https://archive.ics.uci.edu/).
- **The Internet II**: *Web scraping* is akin to a treasure hunt, extracting valuable data directly from web pages, given that you respect the legal and ethical boundaries.
- **Internal Corporate Data**: Picture a gold mine that a company sits upon; these are years' worth of data
- **Academic Data**: Scientifically collected and often meticulously maintained, these datasets can be found accompanying research papers.
- **Data Upon Request**: Sometimes, just asking can open doors.
- **Commercial Datasets**: There are instances when investing in a dataset can prove to be more cost-effective than collecting data from scratch.


# So...you have some data...

Once the data is in your hands, several pivotal steps ensure its quality and relevance:

- **Data Format**: Ensuring that the data is in a readable and workable format is paramount.
- **Source Credibility**: The credibility of the data source can significantly impact the reliability of your findings.
- **Data Processing Knowledge**: A thorough understanding of how the data has been previously processed can save you from potential pitfalls.
- **Legal issues:** Having data is not the same as *being allowed to use it*.


# Example

Imagine you are tasked with evaluating the state of digitization in schools in your region.

**Objective:** Assess the current level of digital infrastructure and its utilization in schools.

***Prompt:*** Spend a few minutes brainstorming:

- What kind of data might you need?
- Where could you potentially source this data?
- Are there potential roadblocks or challenges in procuring or using this data?

# Example (continued)

## What kind of data might you need?

- Number of computers or tablets in each school
- Whether schools have Wi-Fi
- Number of students and teachers
- Types of technology used (laptops, smartboards, projectors)
- Student-to-device ratios
- Internet bandwidth and reliability
- Usage data (how often devices or platforms are used)
- Teacher training or professional development in technology
- Software or learning management systems used
- Hardware availability and age of devices
- Internet speed, uptime, and coverage within schools
- Frequency and purpose of digital tool usage (instructional vs. administrative)
- Equity indicators (access differences by school, neighborhood, or student group)
- Budget allocations for technology and maintenance
- Outcomes related to digitization (e.g., engagement, attendance, performance)

# Example (continued)

## Where could you potentially source this data?

- School websites
- Asking teachers or principals
- School district records
- Government education websites
- School district technology departments
- State or regional education agencies
- Surveys of teachers and students
- Public education reports or audits
- Education department datasets (local, state, or national)
- School IT logs and procurement records
- Vendor reports from educational technology platforms
- Structured surveys and interviews
- Open data portals and academic studies

# Example (continued)

## Potential roadblocks or challenges

- Some schools might not have updated information
- Schools might not want to share data
- Data could be missing or incomplete
- Privacy concerns with student or teacher data
- Differences in how schools define or measure “technology use”
- Data may be spread across multiple systems
- Survey data may be biased or incomplete
- Data privacy laws (e.g., FERPA) limiting access
- Inconsistent data formats across schools or districts
- Difficulty linking technology data to outcomes
- Data that exists but is not publicly accessible
- Potential political or administrative resistance
